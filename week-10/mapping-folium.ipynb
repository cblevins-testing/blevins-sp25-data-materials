```python
def create_map_markers(row, map_name):
    folium.Marker(location=[row['Latitude'], row['Longitude']], popup=row['PO_Name']).add_to(map_name)
```

```python
found_utah_locations = postmasters[postmasters['Latitude'].notna()]
found_utah_locations = postmasters.dropna(subset=['Latitude', 'Longitude'])
```

```python
found_utah_locations
```

```python
found_utah_locations.apply(create_map_markers, map_name=utah_map, axis='columns')
utah_map
```

```python
def create_circles(row, map_name):
    folium.CircleMarker(location=[row['Latitude'], row['Longitude']], radius=20, fill=True,).add_to(map_name)
```

```python
utah_map=folium.Map(location=[39.80664682484737, -111.61123222088217], zoom_start=6)

create_circles(single_pm, utah_map)
```
