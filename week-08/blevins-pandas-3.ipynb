```python
import pandas as pd
```

```python
utah_df=pd.read_csv('utah-census-1880.csv')
utah_df
```

```python
pd.set_option('display.max_columns', None)
utah_df.sample(12)
```

```python
utah_df=utah_df.rename(columns={"birthstate":"birthplace"})
utah_df
```

```python
def replace_married(some_string):
    if some_string=='m':
        return "married"
    elif some_string=='s':
        return 'single'
    elif some_string=='w':
        return 'widowed'
replace_married('s')
```

```python
utah_df['marrystatus']=utah_df['marrystatus'].apply(replace_married)
```
